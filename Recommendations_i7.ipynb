```python
import pandas as pd

file_path = '/content/drive/MyDrive/Sentiment analysis i7/Aspect_based_result_i7.csv'
df = pd.read_csv(file_path)

print("Dataframe head:\n", df.head())

aspect_columns = ['cpu', 'speed', 'processor', 'upgrade', 'core', 'temperature', 'cost']

#calculating averages
average_sentiment_scores = df[aspect_columns].mean().reset_index()
average_sentiment_scores.columns = ['aspect', 'avg_sentiment_score']
print("\nAverage Sentiment Scores:\n", average_sentiment_scores)

negative_aspects = average_sentiment_scores[average_sentiment_scores['avg_sentiment_score'] < 0.5]
print("\nNegative Aspects:\n", negative_aspects)

negative_feedback = df[['reviewDescription', 'classification'] + list(negative_aspects['aspect'])]
negative_feedback = negative_feedback[negative_feedback['classification'] == 'NEGATIVE']

negative_feedback_summary = {}
for aspect in negative_aspects['aspect']:
    feedbacks = negative_feedback[negative_feedback[aspect] > 0.5]['reviewDescription'].tolist()
    negative_feedback_summary[aspect] = feedbacks

# Create a dictionary
summary = {}

for aspect, feedbacks in negative_feedback_summary.items():
    summary[aspect] = {
        "feedback": feedbacks,
        "recommendations": []
    }

summary['cpu']['recommendations'] = [
    "Provide a cooling system to reduce the heat produced.",
    "Provide detailed troubleshooting steps in the user manual."
]

summary['speed']['recommendations'] = [
    "Speed is quite good. As the speed increases there is possibility of overheating.",
    "Consider hardware upgrades to improve processing speed."
]

summary['processor']['recommendations'] = [
    "Improve processor efficiency through firmware updates.",
    "Offer guidelines for optimal processor usage."
]

summary['core']['recommendations'] = [
    "Ensure all cores are utilized efficiently during high-demand tasks.",
    "Provide detailed specifications on core performance."
]

summary['cost']['recommendations'] = [
    "Reevaluate pricing to ensure it aligns with perceived value.",
    "Consider offering discounts or bundles to attract more customers."
]

# Display summary
for aspect, details in summary.items():
    print(f"\n## {aspect.capitalize()}")
    print("### Feedback:")
    for feedback in details['feedback']:
        print(f"- {feedback}")
    print("\n### Recommendations:")
    for recommendation in details['recommendations']:
        print(f"- {recommendation}")
```
